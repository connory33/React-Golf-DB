```python
# Consolidating all schedules to one table
import requests
import pandas as pd

# Unified lists
seasons = []
course_names = []
course_keys = []
event_ids = []
event_names = []
latitudes = []
longitudes = []
start_dates = []
tours = []

# List of tours you want to include
tours_to_fetch = ['pga', 'euro', 'kft', 'liv']

# API key
api_key = '8c29c8197a3391551e4092d3dd9b'

# Loop through each tour
for tour in tours_to_fetch:
    url = f'https://feeds.datagolf.com/get-schedule?tour={tour}&file_format=json&key={api_key}'
    data = requests.get(url).json()
    
    for event in data['schedule']:
        seasons.append(data['current_season'])
        course_names.append(event.get('course'))
        course_keys.append(event.get('course_key'))
        event_ids.append(event.get('event_id'))
        event_names.append(event.get('event_name'))
        latitudes.append(event.get('latitude'))
        longitudes.append(event.get('longitude'))
        start_dates.append(event.get('start_date'))
        tours.append(tour)

# Create a single DataFrame
df = pd.DataFrame({
    'tour': tours,
    'season': seasons,
    'course_name': course_names,
    'course_key': course_keys,
    'event_id': event_ids,
    'event_name': event_names,
    'latitude': latitudes,
    'longitude': longitudes,
    'start_date': start_dates
})

# Export to CSV
df.to_csv('all_tours_events.csv', index=False)
```

```python


event_ids = []; event_completed_date = []; dg_id = []; fin_text = []; birdies = []; bogies = []; course_par = []; doubles_or_worse = []; driving_acc = []; 
driving_dist = []; eagles_or_better = []; gir = []; great_shots = []; pars = []; poor_shots = []; prox_fw = []; prox_rgh = []; score = []; scrambling = []; 
sg_app = []; sg_arg = []; sg_ott = []; sg_putt = []; sg_t2g = []; sg_total = []; start_hole = []; teetime = []; event_round = []

round_keys = ['round_1', 'round_2', 'round_3', 'round_4']
tours = ['pga', 'euro', 'kft', 'cha', 'jpn', 'anz', 'alp', 'champ', 'kor', 'ngl', 'bet', 'chn', 'afr', 'pgt', 'pgti', 'atvt', 'atgt', 'sam', 'ept', 'can', 'liv', 'mex']
# years = ['2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025']
years = ['2025']

url1 = 'https://feeds.datagolf.com/historical-raw-data/rounds?tour='
url2 = '&event_id=all&year='
url3 = '&file_format=json&key=8c29c8197a3391551e4092d3dd9b'

for tour in tours:
    for year in years:
        url = url1 + tour + url2 + year + url3
        print(url)
        if requests.get(url).status_code != 200:
            print(f"Failed to fetch data for {tour} in {year}. Status code: {requests.get(url).status_code}")
            continue
        data = requests.get(url).json()
        for key in data.keys():  # keys are event IDs
            event_scores = data[key]['scores']
            for player in event_scores:
                for i, round_key in enumerate(round_keys, start=1):
                    if round_key in player:
                        round_data = player[round_key]
                        event_ids.append(key)
                        event_completed_date.append(data[key]['event_completed'])
                        dg_id.append(player['dg_id'])
                        fin_text.append(player['fin_text'])
                        event_round.append(i)

                        birdies.append(round_data.get('birdies'))
                        bogies.append(round_data.get('bogies'))
                        course_par.append(round_data.get('course_par'))
                        doubles_or_worse.append(round_data.get('doubles_or_worse'))
                        driving_acc.append(round_data.get('driving_acc'))
                        driving_dist.append(round_data.get('driving_dist'))
                        eagles_or_better.append(round_data.get('eagles_or_better'))
                        gir.append(round_data.get('gir'))
                        great_shots.append(round_data.get('great_shots'))
                        pars.append(round_data.get('pars'))
                        poor_shots.append(round_data.get('poor_shots'))
                        prox_fw.append(round_data.get('prox_fw'))
                        prox_rgh.append(round_data.get('prox_rgh'))
                        score.append(round_data.get('score'))
                        scrambling.append(round_data.get('scrambling'))
                        sg_app.append(round_data.get('sg_app'))
                        sg_arg.append(round_data.get('sg_arg'))
                        sg_ott.append(round_data.get('sg_ott'))
                        sg_putt.append(round_data.get('sg_putt'))
                        sg_t2g.append(round_data.get('sg_t2g'))
                        sg_total.append(round_data.get('sg_total'))
                        start_hole.append(round_data.get('start_hole'))
                        teetime.append(round_data.get('teetime'))

# Create a DataFrame from the lists
df = pd.DataFrame({
    'event_id': event_ids,
    'event_completed_date': event_completed_date,
    'dg_id': dg_id,
    'fin_text': fin_text,
    'event_round': event_round,
    'birdies': birdies,
    'bogies': bogies,
    'course_par': course_par,
    'doubles_or_worse': doubles_or_worse,
    'driving_accuracy': driving_acc,
    'driving_distance': driving_dist,
    'eagles_or_better': eagles_or_better,
    'gir': gir,
    'great_shots': great_shots,
    'pars': pars,
    'poor_shots': poor_shots,
    'prox_fw': prox_fw,
    'prox_rgh': prox_rgh,
    'score': score,
    'scrambling': scrambling,
    'sg_app': sg_app,
    'sg_arg': sg_arg,
    'sg_ott': sg_ott,
    'sg_putt': sg_putt,
    'sg_t2g': sg_t2g,
    'sg_total': sg_total,
    'start_hole': start_hole,
    'teetime': teetime
})
df.head()  
df.to_csv('all_historical_rounds.csv', index=False)

```

```python
# get point and salary info
# url = 'https://feeds.datagolf.com/historical-dfs-data/points?tour=[ tour ]&site=[ site ]&event_id=[ event_id ]&year=[ year ]&file_format=[ file_format ]&key=8c29c8197a3391551e4092d3dd9b'

bogey_free_pts = []
dg_id = []
fin_text = []
finish_pts = []
hole_in_one_pts = []
hole_score_pts = []
ownership = []
player_name = []
salary = []
streak_pts = []
sub_70_pts = []
total_pts = []
event_completed_date = []
event_id = []
event_name = []
ownership_from = []
season = []
site = []
tour = []
calendar_year = []

for row in df.itertuples():
    print(row)
    eventID = row.event_id
    year = row.calendar_year
    url = f'https://feeds.datagolf.com/historical-dfs-data/points?tour=pga&site=draftkings&event_id={eventID}&year={year}&file_format=json&key=8c29c8197a3391551e4092d3dd9b'
    data = requests.get(url)
    if data.status_code != 200:
        print(f"Failed to fetch data for event ID {eventID} in year {year}. Status code: {data.status_code}")
        continue
    data = data.json()

    dfs_points = data['dfs_points'] # dfs point breakdown for each player in the event, includes salary
    for player in dfs_points:
        bogey_free_pts.append(player['bogey_free_pts']) # points for finishing bogey-free
        dg_id.append(player['dg_id']) # DataGolf ID
        fin_text.append(player['fin_text']) # finish text (e.g., "1st", "T2", etc.)
        finish_pts.append(player['finish_pts']) # points for finishing position
        hole_in_one_pts.append(player['hole_in_one_pts']) # points for hole-in-one
        hole_score_pts.append(player['hole_score_pts']) # points for hole scores
        ownership.append(player['ownership']) # ownership percentage
        player_name.append(player['player_name']) # player name
        salary.append(player['salary']) # salary in DFS site currency
        streak_pts.append(player['streak_pts']) # points for streaks (e.g., birdie streak)
        sub_70_pts.append(player['sub_70_pts']) # points for rounds under 70
        total_pts.append(player['total_pts']) # total points scored by the player

        event_completed_date.append(data['event_completed']) # date of event completion
        event_id.append(eventID) # event ID
        event_name.append(data['event_name']) # event name
        ownership_from.append(data['ownerships_from']) # ownership rate
        season.append(data['season']) # season of the event
        site.append(data['site']) # DFS site (e.g., DraftKings, FanDuel)
        tour.append(data['tour']) # tour of the event
        calendar_year.append(data['year']) # year of the event


# Create a DataFrame from the lists
df = pd.DataFrame({
    'bogey_free_pts': bogey_free_pts,
    'dg_id': dg_id,
    'fin_text': fin_text,
    'finish_pts': finish_pts,
    'hole_in_one_pts': hole_in_one_pts,
    'hole_score_pts': hole_score_pts,
    'ownership': ownership,
    'player_name': player_name,
    'salary': salary,
    'streak_pts': streak_pts,
    'sub_70_pts': sub_70_pts,
    'total_pts': total_pts,
    'event_completed_date': event_completed_date,
    'event_id': event_id,
    'event_name': event_name,
    'ownership_from': ownership_from,
    'season': season,
    'site': site,
    'tour': tour,
    'calendar_year': calendar_year
})

df.head()
df.to_csv('historical_dfs_points.csv', index=False) 




```
